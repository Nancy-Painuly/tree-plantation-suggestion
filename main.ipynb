```python
import pandas as pd

data = {
    'Location': ['Dehradun', 'Haridwar', 'Nainital', 'Rishikesh', 'Pauri'],
    'Latitude': [30.3256, 29.9457, 29.3919, 30.0869, 30.1536],
    'Longitude': [78.0437, 78.1642, 79.4542, 78.2676, 78.7815],
    'Green_Cover_Percentage': [42, 18, 27, 35, 22],
    'Soil_Type': ['Sandy', 'Clay', 'Loamy', 'Loamy', 'Sandy'],
    'Water_Source_Distance': [1.5, 3.2, 1.1, 2.0, 1.8]
}

df = pd.DataFrame(data)
df.to_csv("green_cover_data.csv", index=False)  # Save for later use
df.head()
```

```python
# Step 1: Load the dataset
df = pd.read_csv("green_cover_data.csv")

# Step 2: Define conditions
recommended = df[
    (df['Green_Cover_Percentage'] < 30) &
    (df['Water_Source_Distance'] < 2.5) &
    (df['Soil_Type'].isin(['Sandy', 'Loamy']))
]

# Step 3: Display recommended locations
print("ðŸŒ³ Recommended Tree Plantation Locations:")
recommended
```

```python
recommended.to_csv("recommended_locations.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Green_Cover_Percentage', data=df, palette='Greens')
plt.axhline(30, color='red', linestyle='--', label='Threshold (30%)')
plt.title("Green Cover Percentage by Location")
plt.xlabel("Location")
plt.ylabel("Green Cover (%)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Longitude', y='Latitude', data=recommended,
    hue='Soil_Type', size='Green_Cover_Percentage', sizes=(50, 200), palette='Set2'
)
plt.title("Recommended Plantation Spots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
